```python
import pandas as pd
import numpy as np
```

```python
# Q1. Trong số các khách hàng có nhiều hơn một đơn hàng, trung vị số ngày giữa hai lần
# mua liên tiếp (inter-order gap) xấp xỉ là bao nhiêu? (Tính từ orders.csv)

# Đọc dữ liệu từ orders.csv
orders = pd.read_csv('../data/raw/orders.csv')
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders = orders.sort_values(by=['customer_id', 'order_date'])
orders['inter_order_gap'] = orders.groupby('customer_id')['order_date'].diff().dt.days

# Lọc ra các khách hàng có nhiều hơn một đơn hàng
multi_order_customers = orders.groupby('customer_id').filter(lambda x: len(x) > 1)
# Tính trung vị số ngày giữa hai lần mua liên tiếp
median_inter_order_gap = multi_order_customers['inter_order_gap'].median()

print(f"Trung vị số ngày giữa hai lần mua liên tiếp (inter-order gap) là: {median_inter_order_gap} ngày")
```

```python
# Q2. Phân khúc sản phẩm (segment) nào trong products.csv có tỷ suất lợi nhuận gộp
# trung bình cao nhất, với công thức (price − cogs)/price?

# Đọc dữ liệu từ products.csv
products = pd.read_csv('../data/raw/products.csv')
products["gross_profit_margin"] = (products["price"] - products["cogs"]) / products["price"]

# Tính tỷ suất lợi nhuận gộp trung bình cho mỗi phân khúc sản phẩm
products_grouped = products.groupby('segment')['gross_profit_margin'].mean()

# Tìm phân khúc sản phẩm có tỷ suất lợi nhuận gộp trung bình cao nhất
best_segment = products_grouped.idxmax()
best_margin = products_grouped.max()

print(f"Phân khúc sản phẩm có tỷ suất lợi nhuận gộp trung bình cao nhất là: {best_segment} với tỷ suất lợi nhuận gộp trung bình là: {best_margin:.2f}")
```

```python
# Q3. Trong các bản ghi trả hàng liên kết với sản phẩm thuộc danh mục Streetwear (join
# returns với products theo product_id), lý do trả hàng nào xuất hiện nhiều nhất?

# Đọc dữ liệu từ returns.csv
returns = pd.read_csv('../data/raw/returns.csv')

# Kết hợp dữ liệu trả hàng với sản phẩm để lấy thông tin về danh mục
returns_with_products = pd.merge(returns, products, on='product_id', how='left')
general_reason = returns_with_products["return_reason"].value_counts().idxmax()

print(f"Lý do trả hàng xuất hiện nhiều nhất trong các bản ghi trả hàng liên kết với sản phẩm thuộc danh mục Streetwear là: {general_reason}")
```

```python
# Q4. Trong web_traffic.csv, nguồn truy cập (traffic_source) nào có tỷ lệ thoát trung
# bình (bounce_rate) thấp nhất trên tất cả các ngày xuất hiện nguồn đó trong cột traffic_source?
# Đọc dữ liệu từ web_traffic.csv
web_traffic = pd.read_csv('../data/raw/web_traffic.csv')
web_traffic_grouped = web_traffic.groupby('traffic_source')['bounce_rate'].mean()
best_traffic_source = web_traffic_grouped.idxmin()

print(f"Nguồn truy cập có tỷ lệ thoát trung bình thấp nhất là: {best_traffic_source} với tỷ lệ thoát trung bình là: {web_traffic_grouped.min()}")
```

```python
# Q5. Tỷ lệ phần trăm các dòng trong order_items.csv có áp dụng khuyến mãi (tức là promo_id
# không null) xấp xỉ là bao nhiêu?
# Đọc dữ liệu từ order_items.csv
order_items = pd.read_csv('../data/raw/order_items.csv')
order_items["has_promo"] = order_items["promo_id"].notnull() | order_items["promo_id_2"].notna()
promo_percentage = order_items["has_promo"].mean() * 100

print(f"Tỷ lệ phần trăm các dòng trong order_items.csv có áp dụng khuyến mãi là: {promo_percentage:.2f}%")
```

```python
# Q6. Trong customers.csv, xét các khách hàng có age_group khác null, nhóm tuổi nào có số
# đơn hàng trung bình trên mỗi khách hàng cao nhất? (tổng số đơn / số khách hàng trong
# nhóm)

#Đọc dữ liệu từ customers.csv
customers = pd.read_csv('../data/raw/customers.csv')
orders_grouped = orders.groupby("customer_id")["order_id"].count()
orders_with_customers = pd.merge(orders_grouped, customers, on='customer_id', how='left')
orders_with_customers_grouped = orders_with_customers.groupby("age_group")["order_id"].mean()
best_age_group = orders_with_customers_grouped.idxmax()
    
print(f"Nhóm tuổi có số đơn hàng trung bình trên mỗi khách hàng cao nhất là: {best_age_group} với số đơn hàng trung bình là: {orders_with_customers_grouped.max():.2f}")
```

```python
# Q7. Vùng (region) nào trong geography.csv tạo ra tổng doanh thu cao nhất trong
# sales_train.csv?
# Đọc dữ liệu từ geography.csv và sales.csv
orders_items = pd.read_csv('../data/raw/order_items.csv')
geography = pd.read_csv('../data/raw/geography.csv')
returns = pd.read_csv('../data/raw/returns.csv')

#Join orders với order_items để lấy thông tin về doanh thu cho mỗi đơn hàng
orders_with_items = pd.merge(orders, order_items, on='order_id', how='left')
orders_with_items["revenue"] = orders_with_items["unit_price"] * orders_with_items["quantity"]

#Join orders_with_items với geography để lấy thông tin về vùng
orders_with_geography = pd.merge(orders_with_items, geography, on='zip', how='left')

# Tính tổng doanh thu cho mỗi vùng
revenue_by_region = orders_with_geography.groupby("region")["revenue"].sum()

#Tinh so tien da tra cho khach hang theo region
orders_with_geography_with_returns = pd.merge(returns, orders_with_geography, on='order_id', how='left')
returns_by_region = orders_with_geography_with_returns.groupby("region")["refund_amount"].sum()
net_revenue_by_region = revenue_by_region - returns_by_region


best_region = net_revenue_by_region.idxmax()
print(f"Vùng tạo ra tổng doanh thu cao nhất là: {best_region} với tổng doanh thu là: {net_revenue_by_region.max():.2f}")
```

```python
# Q8. Trong các đơn hàng có order_status = ’cancelled’ trong orders.csv, phương thức
# thanh toán nào được sử dụng nhiều nhất?

order_cancelled = orders[orders["order_status"] == "cancelled"]
most_common_payment_method = order_cancelled["payment_method"].value_counts().idxmax()
print(f"Phương thức thanh toán được sử dụng nhiều nhất trong các đơn hàng bị hủy là: {most_common_payment_method}")
```

```python
# Q9. Trong bốn kích thước sản phẩm (S, M, L, XL), kích thước nào có tỷ lệ trả hàng cao
# nhất, được định nghĩa là số bản ghi trong returns chia cho số dòng trong order_items (join
# với products theo product_id)?

#Join orders_items với products để lấy thông tin về kích thước
order_items_with_products = pd.merge(order_items, products, on='product_id', how='left')
#Số orders theo từng size
orders_by_size = order_items_with_products.groupby('size')['order_id'].count()

#Join returns với products để lấy thông tin về kích thước
returns_with_products = pd.merge(returns, products, on='product_id', how='left')
#Số returns theo từng size
returns_by_size = returns_with_products.groupby('size')['return_id'].count()

#Tính tỷ lệ trả hàng theo từng size
return_rate_by_size = returns_by_size / orders_by_size
best_size = return_rate_by_size.idxmax()
print(f"Kích thước có tỷ lệ trả hàng cao nhất là: {best_size} với tỷ lệ trả hàng là: {return_rate_by_size.max():.2f}")

```

```python
# Q10. Trong payments.csv, kế hoạch trả góp nào có giá trị thanh toán trung bình trên
# mỗi đơn hàng cao nhất?
# Đọc file payments.csv để lấy thông tin về phương thức thanh toán
payments = pd.read_csv('../data/raw/payments.csv')
payments_grouped = payments.groupby('installments')['payment_value'].mean()
best_installment_plan = payments_grouped.idxmax()
print(f"Kế hoạch trả góp có giá trị thanh toán trung bình trên mỗi đơn hàng cao nhất là: {best_installment_plan} với giá trị thanh toán trung bình là: {payments_grouped.max():.2f}")
```
